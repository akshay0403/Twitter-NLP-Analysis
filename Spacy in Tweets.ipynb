## Importing Libraries

```python
import pandas as pd
import spacy
import random
from collections import Counter
import seaborn as sns
```

## Loading Spacy 'English' Model

```python
nlp = spacy.load('en_core_web_sm')
```

## Loading the Dataset

```python
tw = pd.read_csv(r"C:\Users\AKSHAY\OneDrive\Desktop\Code\Natural Language Processing\Spacy\Project -- Nike Tweet Analysis\justdoit_tweets_2018_09_07_2.csv")
```

## Taking some Sample Tweets

```python
random.seed(888)
text = tw.tweet_full_text[random.sample(range(1, 100), 10)]
text
```

## NLP Process

- Coverting all the text into ' String ' format

```python
text_combined = str(text)
```

- Converting text to doc

```python
doc = nlp(text_combined)
```

### Tokenization

```python
for token in doc:
    print(token)
```

### Printing Parts of Speech

```python
for token in doc:
    print(token.text, token.pos_)
```

### Printing only Nouns

```python
nouns = list(doc.noun_chunks)
nouns
```

### Tokenization based on Setences

```python
list(doc.sents)
```

### Named Entity Recognition

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

### Visualizing Named Entity

```python
spacy.displacy.render(doc, style = 'ent', jupyter = True)
```

### Lemmatization

```python
for token in doc:
    print(token.text, token.lemma_)
```

### Dependency Parser Visualization

```python
spacy.displacy.render(doc, style = 'dep', jupyter = True)
```
